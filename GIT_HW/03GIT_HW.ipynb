```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

# Question 3.2.

## (A)

```python
N2 = 100
m2 = np.random.rand(N2)
print m2.shape
```

```python
print m2
```

## (B)

$$ G = \left(
    \begin{array}{cccccc}
      1 & 0 & 0 & 0 & 0 & \ldots & 0 \\
      1 & 1 & 0 & 0 & 0 & \ldots & 0 \\
      1 & 1 & 1 & 0 & 0 & \ldots & 0 \\
      0 & 1 & 1 & 1 & 0 & \ldots & 0 \\
      \vdots & \ddots & \ddots & \ddots & \ddots & \ddots & \vdots \\
      0 & \ldots & 0 & 0 & 1 & 1 & 1
    \end{array}
  \right) $$

```python
G2 = np.zeros((N2, N2))

for j in range(N2):
    if j == 0:
        G2[j, j] = 1
    elif j == 1:
        G2[j, j-1] = 1
        G2[j, j] = 1
    else:
        G2[j, j-2] = 1
        G2[j, j-1] = 1
        G2[j, j] = 1
```

```python
print G2
```

## (C)

```python
mn = 0.
std = 1e-2
n2 = np.random.normal(mn, std, size=N2)

print n2.shape
```

```python
d2_obs = np.dot(G2, m2) + n2

print d2_obs
```

## (D)

The least square solution for the model parameter is:

$$ m^{est} = \big[ G^T G \big]^{-1} G^T \ d^{obs} $$

```python
m2_est = np.dot(np.dot(np.linalg.inv(np.dot(G2.T, G2)), G2.T), d2_obs)
```

```python
print m2_est
```

## (E)

Since the measurements of weights should not be correlated, the covariance matrix of the model parameters can be derived as:

$$ cov[m] = {\sigma_d}^2 \big[ G^T G \big]^{-1} $$

```python
cov_m2 = std**2 * np.linalg.inv( np.dot(G2.T, G2) )
```

```python
print cov_m2
```

```python
fig = plt.figure(figsize=(6,5))
ax = fig.add_subplot(111)
ax.set_title('Plot of the covariance matrix of $m$', fontsize=15)
ax.set_ylabel('Rows', fontsize=14)
ax.set_xlabel('Columns', fontsize=14)
cax = ax.imshow(cov_m2)
cbar = fig.colorbar(cax)
plt.xticks( fontsize=12 )
plt.yticks( fontsize=12 )
```

```python
var_m2 = [ cov_m2[i][i] for i in range(N2) ]   # Extract diagonal elements from the covariance matrix of m

print var_m2
```

## (F)

```python
std_m2 = np.sqrt(var_m2)

print std_m2
```

```python
in_2std2 = np.abs( (m2 - m2_est).copy() )
in_2std2[ in_2std2 > 2*std_m2 ] = np.nan  # Convert elements below the two standard deviations from the mean to nans

print in_2std2
```

```python
num_in_2std2 = np.count_nonzero(~np.isnan(in_2std2))  #Count number of non-nan valued elements
print 'The number of the model parameters within two standard deviations is: ', num_in_2std2, ', and this accounts for', num_in_2std2, '% of the total number of the model parameters'
```

For a linear algebra operation, if the operand follows a normal Gaussian distribution, so should the end result. Since the errors in $d_{obs}$ are Gaussian, the model parameters should also have a Gaussian distribution. Hence, the percentage of the model parameters being within $\pm 2 \sigma_m$ of their true value should be close to 95%, which is what we see above.

## (G)

```python
index2 = range(1, N2+1)

fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(111)
ax.set_title('Plot of the standard deviation of $m$', fontsize=15)
n = ax.plot(index2, std_m2)
plt.xlabel('index', fontsize=13)
plt.ylabel('$\sigma_m$', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```

The standard deviation, $\sigma_m$ grows as the indices of the model parameter increases.
